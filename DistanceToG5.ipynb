```python
from astroquery.vizier import Vizier
```

```python
from astropy import coordinates, units as u
```

```python
g5_center = coordinates.SkyCoord(5.4*u.deg, -0.4*u.deg, frame='galactic')
g5_center
```

```python
rslt = Vizier(row_limit=1e5).query_region(g5_center, radius=9*u.arcmin, catalog='I/347')
tbl = rslt[0]
tbl[:2]
```

```python
%matplotlib inline
import pylab as pl
pl.style.use('dark_background')
```

```python
pl.figure(figsize=(20,20))
ax = pl.gca()
sc = ax.scatter(tbl['RA_ICRS'], tbl['DE_ICRS'], alpha=1, s=5, c=tbl['rest'])
ax.set_aspect(1)
pl.colorbar(mappable=sc)
```

```python
pl.figure(figsize=(18,18))
ax = pl.gca()
sel = tbl['rest'] > 4000
sc = ax.scatter(tbl['RA_ICRS'][sel], tbl['DE_ICRS'][sel], alpha=1, s=5, c=tbl['rest'][sel])
ax.set_aspect(1)
pl.colorbar(mappable=sc)
```
