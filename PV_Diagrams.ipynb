```python
def plot_position_velocity(pvdiagram, aspect):
    norm = simple_norm(pvdiagram.data, 'asinh')
    fig = pl.figure(figsize=(6,6))
    ax = pl.subplot(111, projection=wcs.WCS(pvdiagram.header))
    
    #im = ax.imshow(pvdiagram.data, norm=norm, cmap='afmhot')
    im = ax.imshow(pvdiagram.data, norm=norm, cmap='inferno')
    #im = ax.imshow(pvdiagram.data, norm=norm)
    
    ax0 = ax.coords[0]
    ax0.set_format_unit(u.arcmin)
    ax1 = ax.coords[1]
    ax1.set_format_unit(u.km/u.s)
    ax.set_ylabel("Velocity [km/s]")
    #ax.set_xlabel("Offset in Longtitude from $\ell=5.48^{\circ}$ [arcmin]")
    ax.set_xlabel("Offset in Longtitude from $\ell=5.48^{\circ}$ [arcmin]")
    ax.set_aspect(aspect)
    cb = pl.colorbar(mappable=im)
    #cb.set_label("Brightness Temperature [K]")
    cb.set_label("K")
    #cb.set_label(pvdiagram.unit)
    #pl.tight_layout()
```

```python
def plot_maxpv(cube, aspect):
    posv = cube.max(axis=1)#np.max(cube_27.unitless, axis=1)

    pl.figure(figsize=(6,6))
    ww = wcs.WCS(posv.header)
    ax = pl.subplot(111, projection=ww)
    norm = simple_norm(posv.data, 'asinh')
    im = ax.imshow(posv.data, norm=norm)
    
    ax0 = ax.coords[0]
    ax0.set_format_unit(u.arcmin)
    ax1 = ax.coords[1]
    ax1.set_format_unit(u.km/u.s)
    ax.set_ylabel("Velocity [km/s]")
    ax.set_xlabel("Offset in Longtitude from $\ell=5.48^{\circ}$ [arcmin]")
    cb = pl.colorbar(mappable=im)
    cb.set_label(cube.unit)
    
    ax.set_aspect(aspect)
```
